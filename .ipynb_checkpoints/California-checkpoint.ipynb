```python
def snsPairGrid(df):

    ''' Plot a Seaborn Pairgrid w/ KDE & scatter plot of df features'''
    g = sns.PairGrid(df, diag_sharey=False)
    g.fig.set_size_inches(14, 13)
    g.map_diag(sns.kdeplot, lw=2) # draw kde approximation on the diagonal
    g.map_lower(sns.scatterplot, s=15,edgecolor="k", linewidth=1, alpha=0.4) # scattered plot on lower half
    g.map_lower(sns.kdeplot, cmap='plasma', n_levels=5) # kde approximation on lower half
    # decrease n_levels for faster plot
    plt.tight_layout()
    
tlist = ['median_income','total_rooms','housing_median_age','latitude','median_house_value','population']
snsPairGrid(house_df[tlist])
```

### Base Model

- The simplest base model that we can use is the mean of y_train. We will also do a simple linear regressor. Note that this is not appropriate for our data, since there is multicollinearity already.

```python
def get_scores(predictions, true):
    MAE = np.mean(abs(predictions-true))
    RMSE = np.sqrt(np.mean((predictions-true)**2))
    R2 = 1 - np.sum((predictions-true)**2)/np.sum((true-np.mean(true))**2)
    
    return MAE, RMSE, R2
```

```python
from sklearn.model_selection import train_test_split

X, y = house_df.drop(['median_house_value', 'ocean_proximity'], axis=1), house_df['median_house_value']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

- The results for the simplest base model are given below:

```python
dict = {}
mean_pred = y_train.mean()

dict['Mean'] = get_scores([mean_pred for _ in range(len(y_test))], y_test)
dict['Mean'] 
```

- Let us use a linear regression for the simple dataset that we have:

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)

y_pred = reg.predict(X_test)

dict['Linear Regressor'] = get_scores(y_pred, y_test)
dict['Linear Regressor']
```

```python
print(reg.score(X_train, y_train))
print(reg.score(X_test, y_test))
```

- This is already much better than the very rough estimate. But we will try to do even better than this!
- Recall also that there was multicollinearity in the data, so one of the assumptions of linear regression is not satisfied. Let us also check the other assumptions:

```python
y_pred = reg.predict(X_test)
residuals = y_pred - y_test
```

```python
# PLot different diagnostic plots
#plt.rcParams["figure.figsize"] = (20,15)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,6))  
plt.style.use('seaborn')
  
# Residual vs Fitted Plot
sns.scatterplot(x=y_pred, y=residuals, ax=ax[0])
ax[0].axhline(y=0, color='grey', linestyle='dashed')
ax[0].set_xlabel('Fitted Values')
ax[0].set_ylabel('Residuals')
ax[0].set_title('Residuals vs Fitted values')
  
# Normal Q-Q plot
sm.qqplot(residuals, fit=True, line='45',ax=ax[1], c='#4C72B0')
ax[1].set_title('Normal Q-Q')
  
```

1. We see clear pattern in residual vs fitted values.
2. Residuals do not follow normal distribution.

So there is room for improvement! It is clear that we should not use a linear model given the raw data.

### K-means clustering

Geography can play an important role in deciding the price of a house. As a result, we can try to use the latitude and longitude in a KMeans clustering algorithm. We will also add the water_proximity variable.

```python
from sklearn.cluster import KMeans

features = ['longitude', 'latitude', 'water_proximity']

kmeans = KMeans(n_clusters = 10)
house_df1['Geographical Clusters'] = kmeans.fit_predict(house_df1[features])
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=house_df1, hue="Geographical Clusters", palette="coolwarm")
```

- Is this value of k good? Let us try to use the elbow method to determine if so.

```python
def calculate_WSS(points, kmax):
    sse = []
    for k in range(1,kmax):
        kmeans = KMeans(n_clusters = k).fit(points)
        centroids = kmeans.cluster_centers_
        pred_clusters = kmeans.predict(points)
        curr_sse = 0
        
        for i in range(len(points)):
            curr_center = centroids[pred_clusters[i]]
            curr_sse += sum((points.iloc[i] - curr_center)**2)
        sse.append(curr_sse)
    return sse
```

```python
features = ['longitude', 'latitude', 'water_proximity']
s0 = calculate_WSS(house_df1[features], 15)
```

```python
plt.plot([k for k in range(1,15)], s0)
```

- Looks like k=5 might already be enough, let's see:

```python
from sklearn.cluster import KMeans

features = ['longitude', 'latitude', 'water_proximity']

kmeans = KMeans(n_clusters = 7)
house_df1['Geographical Clusters'] = kmeans.fit_predict(house_df1[features])

plt.figure(figsize=(15,8))
sns.scatterplot(x="latitude", y="longitude", data=house_df1, hue="Geographical Clusters", palette="coolwarm")
```
